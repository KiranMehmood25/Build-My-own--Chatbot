```python
from tkinter import *

wind = Tk()
wind.title('ChatBot')
wind.geometry('2000x2000')
wind.configure(bg='#0084ff')
xi = 0
yi = 0

chat_history = []

def clear_message():
    global yi
    for widget in chat_frame.winfo_children():
        widget.destroy()
    yi = 0

def save_chat_history():
    with open("chat_history.txt", "w") as file:
        for message in chat_history:
            file.write(message + "\n")

def send_message():
    global yi
    U = user_entry.get()
    user = Label(chat_frame, text=U + ' <<You', height=1, width=146, bg='#a6a6a6', fg='black', font=12, anchor='e')
    user.grid(row=yi, column=0, sticky='e')

    if 'hello' in U:
        bot_text = 'bot>>Hello'
    elif 'how are you?' in U:
        bot_text = 'bot>>Im fine'
    
    elif 'tell me a joke' in U:
        bot_text = 'bot>>Why did the computer keep freezing? Because it left its Windows open!'
    elif 'thank you' in U:
        bot_text = 'bot>>You\'re welcome!'
    elif 'you have information about python programming language' in U:
        bot_text = 'bot>>yes whynot you can ask feel free about python'
    elif 'goodbye' in U:
        bot_text = 'bot>>Goodbye! Have a great day!'
    elif 'help' in U:
        bot_text = 'bot>>I can answer questions, tell jokes, or chat with you about Python Language. Just ask!'
    elif 'who is your developer?' in U:
        bot_text = 'bot>>My developer is a kiran Mehmood she is very talented programmer who created me.'
    elif 'tell me something about kiran?' in U:
        bot_text = 'bot>>yes she is the best developer in the world and she also create me.'
    elif 'what is your name?' in U:
        bot_text = 'bot>>My name is PYCHAT.'
    elif 'who are you?' in U:
        bot_text = 'bot>>I am a chatbot created to assist and chat with you.'
    elif 'how old are you?' in U:
        bot_text = 'bot>>I don\'t have an age, I\'m a program.'
    elif 'what can you do?' in U:
        bot_text = 'bot>>I can provide information, answer questions, tell jokes, and more.'
    elif 'where are you from?' in U:
        bot_text = 'bot>>I exist in the digital world, so I don\'t have a physical location.'
    elif 'tell me a fun fact' in U:
        bot_text = 'bot>>Did you know that honey never spoils? Archaeologists have found pots of honey in ancient Egyptian tombs that are over 3,000 years old and still perfectly edible!'
    
       
    elif 'what is python?' in U:
        bot_text = 'bot>>Python is a high-level, interpreted programming language.'
    elif 'How do you print "Hello, World!" in Python?' in U:
        bot_text = 'bot>>Use print("Hello, World!").'
    elif 'what is a variable in python?' in U:
        bot_text = 'bot>>A named container to store data.'
    elif 'how do you declare a variable in python?' in U:
        bot_text = 'bot>>Assign a value, like x = 10.'
    elif 'what are data types in python?' in U:
        bot_text = 'bot>>Integer, float, string, list, tuple, dictionary, etc.'
    elif 'How do you perform arithmetic operations in Python?' in U:
        bot_text = 'bot>>Use +, -, *, /, etc., e.g., result = 5 + 3.'
    elif 'what is a conditional statement in python?' in U:
        bot_text = 'bot>>Used for decision-making, like if, elif, and else.'
    elif 'Provide an example of an if statement in Python.' in U:
        bot_text = 'bot>>if age >= 18: print("You are an adult.")'
    elif 'what are loops in python and how are they used?' in U:
        bot_text = 'bot>>Loops repeat code; for and while are common.'
    elif 'Give an example of a for loop in Python.' in U:
        bot_text = 'bot>>for i in range(1, 6): print(i)'
    elif 'what is a function in python?' in U:
        bot_text = 'bot>>A reusable block of code defined with def.'
    elif 'How do you define and call a function in Python?' in U:
        bot_text = 'bot>>Define: def greet(name): Call: greet("Alice").'
    elif 'What is a list in Python?' in U:
        bot_text = 'bot>>Ordered collection of items.'
    elif 'Provide an example of creating and manipulating a list in Python.' in U:
        bot_text = 'bot>>[1, 2, 3], append(4), modify with indexing.'
    elif 'What is a dictionary in Python?' in U:
        bot_text = 'bot>>Unordered key-value pairs collection.'
    elif 'Show how to create and access values in a dictionary in Python.' in U:
        bot_text = 'bot>>{"name": "Alice", "age": 30}, access with keys.'
    elif U == 'clear':
        clear_message()
        user_entry.delete(0, 'end')
        return
    else:
        bot_text = 'bot>Sorry, I do not have a response for that'


        
    bot_message = Label(chat_frame, text=bot_text, height=1, width=146, bg='white', fg='black', font=12, anchor='w')
    bot_message.grid(row=yi + 1, column=0, sticky='w')

    chat_history.append(U + ' <<You')
    chat_history.append(bot_text)

    yi += 2  
    user_entry.delete(0, 'end')

    
    chat_canvas.update_idletasks() 
    chat_canvas.config(scrollregion=chat_canvas.bbox("all"))

# Create vertical scrollbar
y_scrollbar = Scrollbar(wind, orient=VERTICAL)
y_scrollbar.pack(side=RIGHT, fill=Y)

# Create horizontal scrollbar
x_scrollbar = Scrollbar(wind, orient=HORIZONTAL)
x_scrollbar.pack(side=BOTTOM, fill=X)

hcb_text = Label(height=1, width=60, bg='#0084ff', text='<PYCHAT>', font=('Impact', 20), fg='white')
hcb_text.place(x=270, y=5)

# Create chat canvas with vertical and horizontal scrollbars
chat_canvas = Canvas(wind, height=520, width=1320, bg='#f5f5f5', yscrollcommand=y_scrollbar.set, xscrollcommand=x_scrollbar.set)
chat_canvas.place(x=10, y=60)

# Create a frame to hold chat messages
chat_frame = Frame(chat_canvas, bg='white',width=1000)
chat_canvas.create_window((0, 0), window=chat_frame, anchor='nw')

# Configure vertical scrollbar for chat_canvas
y_scrollbar.config(command=chat_canvas.yview)
# Configure horizontal scrollbar for chat_canvas
x_scrollbar.config(command=chat_canvas.xview)

entry_bg = Frame(height=80, width=700, bg='white')
entry_bg.place(x=10, y=600)

sendbtn_bg = Frame(height=80, width=635, bg='white')
sendbtn_bg.place(x=700, y=600)

def on_enter(e):
    user_entry.delete(0, 'end')
    user_entry.config(fg='black')

def on_leave(e):
    n = user_entry.get()
    user_entry.config(fg='#5c5a5a')
    if n == '' or n == '':
        user_entry.insert(0, 'Enter message...')
        user_entry.config(fg='#5c5a5a')

user_entry = Entry(entry_bg, width=32, bg='white', font=('Helvetica', 20), relief=FLAT, border=0)
user_entry.place(x=10, y=20)
user_entry.insert(0, 'Enter message...')
user_entry.config(fg='#5c5a5a')
user_entry.bind("<FocusIn>", on_enter)
user_entry.bind("<FocusOut>", on_leave)

# Bind the <Return> key to the send_message function
user_entry.bind("<Return>", lambda event=None: send_message())

send_button = Button(sendbtn_bg, height=1, width=3, bg='#0084ff', text='>', font=('Helvetica', 20),
                    activeforeground='white', fg='white', relief=FLAT, border=0,
                    activebackground='#0084ff', command=send_message)
send_button.place(x=560, y=12)

# Button to save chat history
#save_button = Button(wind, text='Save Chat', command=save_chat_history)
#save_button.place(x=10, y=650)

wind.mainloop()
```
